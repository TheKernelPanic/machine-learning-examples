```python
import pandas
import matplotlib.pyplot as pyplot

fruits = pandas.read_csv("datasets/fruits_generated.csv", names=['DIAMETER', 'WEIGHT'], header=None)

fruits.plot.scatter(x="DIAMETER", y="WEIGHT")
pyplot.show()
```

```python
from sklearn.cluster import KMeans

model=KMeans(n_clusters=2)

model.fit(fruits)

kmeans_predictions = model.predict(fruits)

pyplot.scatter(fruits.DIAMETER, fruits.WEIGHT, c=kmeans_predictions, s=50, cmap='viridis')
pyplot.xlabel("DIAMETER")
pyplot.ylabel("WEIGHT")

centers = model.cluster_centers_
pyplot.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
pyplot.show()
```

```python
 from joblib import dump
dump(model, 'models/kmean.joblib')
```
