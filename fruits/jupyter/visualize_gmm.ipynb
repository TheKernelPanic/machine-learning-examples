```python
import pandas
import matplotlib.pyplot as pyplot

fruits = pandas.read_csv("datasets/fruits_generated.csv", names=['DIAMETER', 'WEIGHT'], header=None)

from sklearn import mixture

gmm = mixture.GaussianMixture(n_components=2)

gmm.fit(fruits)

clusters = gmm.predict(fruits)

pyplot.scatter(fruits.DIAMETER, fruits.WEIGHT, c=clusters, s=40, cmap='viridis')
pyplot.xlabel("DIAMETER")
pyplot.ylabel("WEIGHT")
pyplot.show()
```

```python
from joblib import dump
dump(gmm, 'models/gmm.joblib')
```
