```python
import random
import pandas
```

```python
# Diámetro mínimo
# Diametro máximo
# Peso mínimo (g)
# Peso máximo (g)
characteristics_cherries = [
    [17,19,1,5],
    [20,21,5,6],
    [22,23,6,7],
    [24,25,7,8.8],
    [26,27,8.5,10],
    [28,29,10,11.5]
]

# Diámetro mínimo
# Diámetro máximo
# Peso medio
characteristics_apricots = [
    [35,39,27],
    [40,44,41],
    [45,49,54],
    [50,54,74],
    [55,59,100]
]
```

```python
amount_observations = 2000

# Generate cherries
cherries = []
random.seed()
for iteration in range(amount_observations):
    cherry = random.choice(characteristics_cherries)
    diameter = round(random.uniform(cherry[0],cherry[1]), 2)
    weight = round(random.uniform(cherry[2], cherry[3]), 2)
    cherries.append([diameter, weight])

# Generate apricots
apricots = []
random.seed()
for iteration in range(amount_observations):
    apricot = random.choice(characteristics_apricots)
    diameter = round(random.uniform(apricot[0], apricot[1]), 2)
    limit_min_weight = apricot[2] / 1.10
    limit_max_weight = apricot[2] * 1.10
    weight = round(random.uniform(limit_min_weight, limit_max_weight), 2)
    apricots.append([diameter, weight])

fruits = cherries + apricots
random.shuffle(fruits)

data_frame = pandas.DataFrame(fruits)
data_frame.to_csv('datasets/fruits_generated.csv', index=False, header=False)

fruits
```
