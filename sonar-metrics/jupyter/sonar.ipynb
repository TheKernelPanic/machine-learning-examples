```python
names = []
i = 1
while i <= 60:
    names.append("F" + str(i))
    i += 1
names.append("Object")
names
```

```python
observations = pandas.read_csv("../datasets/sonar.all-data.csv", names=names)
```

```python
# Transformar la caracteristica Object
observations['Object'] = (observations['Object'] == 'M').astype(int)
```

```python
from matplotlib import pyplot

observations.plot.box(figsize=(20, 10), xticks=[])

pyplot.title('Detección de los valores extremos')
pyplot.xlabel('Las 60 frecuencias')
pyplot.ylabel('Potencia de la señal')
pyplot.show()

```

```python
# Obtener datos de apredizaje y de validación
from sklearn.model_selection import train_test_split

as_array = observations.values

# Conversión de datos a tipo decimal
x = as_array[:,0:-1].astype(float)

# Utilizar la última columna como caracteristica de predicción
y = as_array[:,-1]

X_LEARN, X_VALIDATION, Y_LEARN, Y_VALIDATION = train_test_split(x, y, test_size=0.2, random_state=42)

X_LEARN
```

```python
# Importar algoritmos de clasificación
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC

```

```python
# Regresión logística
logistic_regression = LogisticRegression()
logistic_regression.fit(X_LEARN, Y_LEARN)
predictions = logistic_regression.predict(X_VALIDATION)
print("Regresión logística: " + str(accuracy_score(predictions, Y_VALIDATION)))
```

```python
# Árbol de decisión
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_LEARN, Y_LEARN)
predictions = decision_tree.predict(X_VALIDATION)
print("Árbol de decisión: " + str(accuracy_score(predictions, Y_VALIDATION)))
```

```python
# Bosque aleatorio
random_forest = RandomForestClassifier()
random_forest.fit(X_LEARN, Y_LEARN)
predictions = random_forest.predict(X_VALIDATION)
print("Bosque aleatorio: " + str(accuracy_score(predictions, Y_VALIDATION)))
```

```python
# K Vecinos más cercanos
knn = KNeighborsClassifier()
knn.fit(X_LEARN, Y_LEARN)
predictions = knn.predict(X_VALIDATION)
print("K vecinos más cercanos: " + str(accuracy_score(predictions, Y_VALIDATION)))
```

```python
# Máquina de vectores de soporte
SVM = SVC(gamma='auto')
SVM.fit(X_LEARN, Y_LEARN)
predictions = SVM.predict(X_VALIDATION)
print("Máquina de vectores de soporte: " + str(accuracy_score(predictions, Y_VALIDATION)))
```
